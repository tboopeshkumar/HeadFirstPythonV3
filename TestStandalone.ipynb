```python
def event_lookup(filename):
    conversions = {
        "Free" : "freestyle",
        "Back": "backstroke",
        "Breast" : "breaststroke",
        "Fly": "butterfly",
        "IM": "individual medley"
    }
    *_, distance, stroke = filename.removesuffix(".txt").split("-")
    return f"{distance} {conversions[stroke]}"
print(event_lookup("Darius-13-100m-IM.txt"))
```
