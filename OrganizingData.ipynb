```python
import os
import swimclub

swim_files = os.listdir(swimclub.FOLDER)
swim_files.remove('.DS_Store')
print(len(swim_files))
```

```python
swimmerslist = []
for file in swim_files:
    swimmerslist.append((swimclub.read_swim_data(file)[0], file))

print(swimmerslist)
```

```python
names = list(set(swimmerslist))

print(names)
```

```python
swimmers = {}
for file in swim_files:
    name, *_ = swimclub.read_swim_data(file)
    if name not in swimmers:
        swimmers[name] = []
    swimmers[name].append(file)
```
