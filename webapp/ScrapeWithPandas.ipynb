```python
import pandas as pd
import requests

URL = "https://en.wikipedia.org/wiki/List_of_world_records_in_swimming"

headers = {
    "User-Agent": "Mozilla/5.0 (compatible; Pandas/2.x; Python)"
}

response = requests.get(URL, headers=headers)
response.raise_for_status()  # raises exception for non-200 responses

tables = pd.read_html(response.text)

```

```python
df =tables[0][["Event", "Time"]]

df.head()

df.tail()
```

```python
df = df[~df["Event"].str.contains("relay")]
```

```python
df.to_dict("records")
```

```python
dfi = df.set_index("Event")
dfi.head()
```

```python
dfi.to_dict()
```

```python
records = {}
records["LC Men"] = dfi.to_dict()["Time"]
```

```python
RECORDS = (0, 1, 3, 4)
COURSES = ("LC Men", "LC Women", "SC Men", "SC Women")

records = {}
for table, course in zip(RECORDS, COURSES):
    df = tables[table][["Event", "Time"]]
    df = df[~df["Event"].str.contains("relay")]
    df = df.set_index("Event")
    records[course] = df.to_dict()["Time"]

print(records)
```

```python
import json

with open("records.json") as jf:
    gazpacho_records = json.load(jf)

records == gazpacho_records
```
