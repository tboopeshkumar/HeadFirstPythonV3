```python
URL = "https://en.wikipedia.org/wiki/List_of_world_records_in_swimming"

import gazpacho

html = gazpacho.get(URL)

len(html)
```

```python
soup = gazpacho.Soup(html)
type(soup)
```

```python
print(dir(soup))
```

```python
tables = soup.find("table")
```

```python
type(tables)
```

```python
for i, table in enumerate(tables, 0):
    table_no = i
    table_rows = table.find("tr", mode="all")
    table_cols = table_rows[-1].find("td", mode="all")    
    print(f"{i} -> {len(table_rows)} rows, {len(table_cols)} cols");
```

```python
table = tables[3]
```

```python
for row in table.find("tr", mode="all")[1:]:
    columns = row.find("td", mode="all")
    event = columns[0].text
    time = columns[1].text
    print(f"{event} -> {time}")
```

```python
RECORDS = (0, 1, 3, 4)
COURSES = ("LC Men", "LC Women", "SC Men", "SC Women")
```

```python
list(zip(RECORDS, COURSES))
```

```python
for table, course in zip(RECORDS, COURSES):
    print(f"{course}:")
    for row in tables[table].find("tr", mode="all")[1:]:
        columns = row.find("td", mode="all")
        event = columns[0].text
        time = columns[1].text
        print(f"\t{event} -> {time}")
    print()
```
